# 1. Problem Statement and Business Understanding

## Business Problem
Many individuals struggle to obtain loans due to limited or insufficient credit history. Financial institutions must balance providing loans to customers while minimizing the risk of default. Incorrect loan approval decisions can lead to significant financial losses for lenders and unfair rejection of reliable applicants. The goal of this project is to develop a machine learning solution that predicts whether a loan applicant is likely to default on a loan using demographic, financial, and credit-related information.

## Project Objective
This project uses the Home Credit Default Risk dataset to analyze customer information and build predictive machine learning models that can classify applicants based on their repayment risk.

Several machine learning techniques will be explored and compared, including:

- Decision Tree
- Random Forest
- XGBoost

The final objective is to identify the model that best balances predictive performance and business interpretability while providing actionable recommendations for financial lending decisions.

## Business Value
A successful prediction model can help financial institutions:

- Reduce loan default risk
- Improve loan approval decisions
- Increase fairness in lending
- Better understand customer risk profiles
- Support data-driven financial decision making

# 2. Dataset Understanding

## Dataset Description
The dataset used in this project is the Home Credit Default Risk dataset (https://www.kaggle.com/c/home-credit-default-risk/data). The dataset contains demographic, financial, and credit-related information about loan applicants.

The primary target variable is:
- TARGET
  - 0 = Client repaid the loan
  - 1 = Client had payment difficulties / default risk

The dataset includes information such as:

- Income
- Credit amount
- Loan annuity
- Employment information
- Family information
- Housing information
- Historical credit behavior
- External credit scores

This dataset is highly suitable for supervised machine learning classification problems because the goal is to predict whether a client is likely to default on a loan.
